# Task 5: Reroll Ability Analysis

This analysis explores the "Reroll" ability. We evaluate:
1. Does the average difficulty (level) of available doors affect the decision to reroll?
2. How does the number of active skills affect the decision to reroll?
3. How do these factors differ between `Control` runs and `Tool` runs?

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Reroll data
with open('../scripts/parsed_task6.json', 'r') as f:
    reroll_data = json.load(f)
df_reroll = pd.DataFrame(reroll_data)

if df_reroll.empty:
    print("No reroll actions found in the current datasets.")
else:
    print(f"Loaded {len(df_reroll)} Reroll events.")
    
    # ==========================================================
    # 1. Compare Control vs Tool: Average Door Level when Rerolling
    # ==========================================================
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df_reroll, x='mode', y='avg_door_level', palette='Set1')
    sns.stripplot(data=df_reroll, x='mode', y='avg_door_level', color='black', alpha=0.5, jitter=True)
    plt.title('Average Level of Available Doors When Rerolling (Control vs Tool)')
    plt.xlabel('Mode')
    plt.ylabel('Average Door Level')
    plt.tight_layout()
    plt.show()

    # ==========================================================
    # 2. Active Skills Count During Reroll (Control vs Tool)
    # ==========================================================
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_reroll, x='active_skills_count', hue='mode', palette='Set3')
    plt.title('Number of Active Skills When Rerolling')
    plt.xlabel('Active Skills Count')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```
