# Task 5: Agent Memory Analysis

This notebook analyzes `parsed_memory.json` to see how often the LLM agent utilizes its memory tool and what it stores on average.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Load parsed memory data
try:
    with open('../scripts/parsed_memory.json', 'r') as f:
        data = json.load(f)
    print(f"Loaded {len(data)} memory events.")
except FileNotFoundError:
    print("No parsed_memory.json found. Check if the memory parser has run.")
    data = []

if not data:
    print("Memory data is empty. The agent may not have utilized the memory tool in these runs.")
else:
    df_mem = pd.DataFrame(data)
    
    # 1. Total memory events by mode
    plt.figure(figsize=(8, 5))
    ax1 = sns.countplot(data=df_mem, x='mode', palette='Set2')
    plt.title('Total Number of Memory State Saves by Mode')
    plt.xlabel('Mode')
    plt.ylabel('Count of Memory Saves')
    for container in ax1.containers:
        ax1.bar_label(container)
    plt.tight_layout()
    plt.show()

    # 2. Average memory items stored per save event
    plt.figure(figsize=(8, 5))
    ax2 = sns.boxplot(data=df_mem, x='mode', y='memory_items_count', palette='Set3')
    # fallback to box/violin or stripplot
    sns.stripplot(data=df_mem, x='mode', y='memory_items_count', color='black', alpha=0.5, jitter=True)
    plt.title('Distribution of Memory Items Count Stored per Save')
    plt.xlabel('Mode')
    plt.ylabel('Number of Memory Items in Store')
    plt.tight_layout()
    plt.show()

    # 3. View some stored samples
    print("\nSample Memory Store Outputs:")
    sample = df_mem.sample(min(5, len(df_mem)))
    display(sample[['run_name', 'chamber_index', 'memory_items_count', 'memory_store']])
```
